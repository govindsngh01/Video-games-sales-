#**Data cleaning Conclusion**


The dataset was checked for missing values before performing analysis.

The Year column contained 271 missing values (~1.6%), which is a very small portion of the dataset. Since the year feature is important for trend analysis, Rows with Null values were removed to maintain data accuracy.

The Publisher column had 58 missing values. As it is a categorical column and the missing count was low, these values were filled with ""Unknown"" instead of dropping rows to avoid unnecessary data loss.

After handling missing values and correcting data types, the dataset became clean and consistent.

No duplicate data , The final dataset is now ready for Exploratory data analysis and visualization.




##EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns
```

```python
## Top 10 best selling video games

top_games=df.sort_values('Global_Sales',ascending=False).head(10)
plt.figure(figsize=(7,6))
sns.barplot(y='Global_Sales', x='Name', data=top_games)
plt.title("Top 10 best selling video games")
plt.xlabel("Game Names")
plt.ylabel("Global sales")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
## Genre wise sales
genre_sales=df.groupby('Genre',as_index=False)['Global_Sales'].sum().sort_values(by='Global_Sales',ascending=False)
plt.figure(figsize=(8,7))
sns.barplot(x='Genre',y='Global_Sales',data=genre_sales)
plt.xticks(rotation=45)
plt.title("Global sales by Genre")
plt.show()
```

```python
## Platform Analysis
platfrom_sales=df.groupby('Platform', as_index=False)['Global_Sales'].sum().sort_values(by='Global_Sales',ascending=False).head(10)
plt.figure(figsize=(7,6))
sns.barplot(x='Platform', y='Global_Sales',data=platfrom_sales)
plt.xticks(rotation=45)
plt.title('Top platforms by global sales')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df[["NA_Sales","EU_Sales","JP_Sales"]])
plt.title("Regional Sales Distribution")
plt.show()
```

```python
Numaric_data = df.select_dtypes(include = ['int64' , 'float64'])


Outlier = []

for col in Numaric_data.columns:
  q1 = np.quantile(df[col] , 0.25)
  q3 = np.quantile(df[col] , 0.75)
  IQR = q3 - q1
  lower_bound = q1 - (1.5 * IQR)
  upper_bound = q3 + (1.5 * IQR)
  Outlier = df[~df[col].between(lower_bound , upper_bound)]

  print(f'{col} : {len(Outlier)}')
```

```python
## Yearly sales trend
year_sales=df.groupby('Year')['Global_Sales'].sum()
plt.figure(figsize=(12,5))
plt.plot(year_sales)
plt.title('Global Video game sales trend by year')
plt.show()
```

```python
## Region wise sales
region=df[['NA_Sales','EU_Sales', 'JP_Sales']].sum()
region.plot(kind='bar')
plt.title("Region wise total sales")
plt.show()
```
